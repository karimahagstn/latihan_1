```python
import networkx as nx
import matplotlib.pyplot as plt

# Membuat graf berbobot
G = nx.Graph()
edges = {
    ('A', 'B'): 12, ('A', 'C'): 8, ('B', 'C'): 7, ('B', 'E'): 15,
    ('C', 'E'): 25, ('C', 'D'): 16, ('E', 'G'): 11, ('E', 'D'): 9,
    ('D', 'F'): 8, ('F', 'G'): 10
}

# Menambahkan edge ke dalam graf
for (u, v), weight in edges.items():
    G.add_edge(u, v, weight=weight)

# Menghitung jalur terpendek dari A ke F menggunakan Dijkstra
shortest_path = nx.dijkstra_path(G, 'A', 'F', weight='weight')
shortest_path_edges = list(zip(shortest_path, shortest_path[1:]))

# Menggambar graf
pos = nx.spring_layout(G, seed=42)  # Posisi node yang tetap
plt.figure(figsize=(8, 6))

# Gambar semua edge dengan warna default
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=2000, font_size=12)

# Gambar edge yang merupakan jalur terpendek dengan warna merah
nx.draw_networkx_edges(G, pos, edgelist=shortest_path_edges, edge_color='red', width=2)

# Menampilkan bobot pada setiap edge
edge_labels = {(u, v): f"{w}" for (u, v), w in edges.items()}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=10, font_color='black')

plt.title("Shortest Path from A to F using Dijkstra")
plt.show()
```
