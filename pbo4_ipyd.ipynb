```python
# Banking System Simulation
class BankAccount:
    def __init__(self, account_number, balance=0):
        self.account_number = account_number
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited ${amount}. New balance: ${self.balance}")

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient funds!")
        else:
            self.balance -= amount
            print(f"Withdrew ${amount}. New balance: ${self.balance}")

    def display(self):
        print(f"Account: {self.account_number}, Balance: ${self.balance}")

class SavingsAccount(BankAccount):
    def withdraw(self, amount):
        if amount > 500:
            print("Withdrawal limit exceeded ($500 max per transaction)")
        else:
            super().withdraw(amount)

class PremiumSavingsAccount(SavingsAccount):
    def withdraw(self, amount):
        if amount > 2000:
            print("Withdrawal limit exceeded ($2000 max per transaction)")
        else:
            super().withdraw(amount)


# Employee Management System
class Employee:
    def __init__(self, name, emp_id, salary):
        self.name = name
        self.emp_id = emp_id
        self.salary = salary

    def calculate_salary(self):
        return self.salary

    def show_details(self):
        print(f"Name: {self.name}, ID: {self.emp_id}, Salary: ${self.calculate_salary()}")

class Manager(Employee):
    def calculate_salary(self):
        return self.salary * 1.2

class Engineer(Employee):
    def calculate_salary(self):
        return self.salary * 1.1

class Intern(Employee):
    def calculate_salary(self):
        return self.salary


# Vehicle Rental System
class Vehicle:
    def __init__(self, brand, model, rental_rate):
        self.brand = brand
        self.model = model
        self.rental_rate = rental_rate

    def calculate_rental(self, days):
        return self.rental_rate * days

class Car(Vehicle):
    def open_trunk(self):
        print("Trunk opened.")

class Bike(Vehicle):
    def kickstart(self):
        print("Bike kickstarted.")

class LuxuryFeatures:
    def enable_gps(self):
        print("GPS enabled.")

    def enable_heated_seats(self):
        print("Heated seats enabled.")

class LuxuryCar(Car, LuxuryFeatures):
    def calculate_rental(self, days):
        return super().calculate_rental(days) + (50 * days)


# Example Usage
if __name__ == "__main__":
    while True:
        print("\n1. Banking System\n2. Employee Management\n3. Vehicle Rental System\n4. Exit")
        choice = input("Enter your choice: ")

        if choice == "1":
            acc_number = input("Enter account number: ")
            acc_balance = float(input("Enter initial balance: "))
            acc1 = PremiumSavingsAccount(acc_number, acc_balance)
            acc1.deposit(float(input("Enter deposit amount: ")))
            acc1.withdraw(float(input("Enter withdrawal amount: ")))
            acc1.display()

        elif choice == "2":
            emp_name = input("Enter employee name: ")
            emp_id = input("Enter employee ID: ")
            emp_salary = float(input("Enter base salary: "))
            emp_role = input("Enter role (Manager/Engineer/Intern): ")

            if emp_role.lower() == "manager":
                emp = Manager(emp_name, emp_id, emp_salary)
            elif emp_role.lower() == "engineer":
                emp = Engineer(emp_name, emp_id, emp_salary)
            else:
                emp = Intern(emp_name, emp_id, emp_salary)

            emp.show_details()

        elif choice == "3":
            brand = input("Enter vehicle brand: ")
            model = input("Enter vehicle model: ")
            rental_rate = float(input("Enter rental rate per day: "))
            vehicle_type = input("Enter vehicle type (Car/Bike/LuxuryCar): ")

            if vehicle_type.lower() == "car":
                vehicle = Car(brand, model, rental_rate)
            elif vehicle_type.lower() == "bike":
                vehicle = Bike(brand, model, rental_rate)
            else:
                vehicle = LuxuryCar(brand, model, rental_rate)

            days = int(input("Enter number of rental days: "))
            print(f"Total rental cost: ${vehicle.calculate_rental(days)}")

            if isinstance(vehicle, Car):
                vehicle.open_trunk()
            if isinstance(vehicle, Bike):
                vehicle.kickstart()
            if isinstance(vehicle, LuxuryCar):
                vehicle.enable_gps()
                vehicle.enable_heated_seats()

        elif choice == "4":
            print("Exiting program...")
            break

        else:
            print("Invalid choice, please try again.")
```
