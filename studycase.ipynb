```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1. Membaca citra asli
image_path = 'aliandosyarif.jpeg'  # Ganti dengan path citra Anda
original_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# 2. Menghitung dan menampilkan histogram dari citra asli
def plot_histogram(image, title):
    plt.hist(image.ravel(), bins=256, range=[0, 256], color='black')
    plt.title(title)
    plt.xlabel('Intensitas Pixel')
    plt.ylabel('Frekuensi')
    plt.xlim([0, 256])
    plt.grid()
    plt.show()

plot_histogram(original_image, 'Histogram Citra Asli')

# 3. Melakukan histogram equalization
equalized_image = cv2.equalizeHist(original_image)

# 4. Menghitung dan menampilkan histogram dari citra hasil equalization
plot_histogram(equalized_image, 'Histogram Citra Hasil Equalization')

# 5. Membandingkan citra asli dan citra hasil equalization
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(original_image, cmap='gray')
plt.title('Citra Asli')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(equalized_image, cmap='gray')
plt.title('Citra Hasil Equalization')
plt.axis('off')

plt.show()
```
