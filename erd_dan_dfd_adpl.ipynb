```python
import matplotlib.pyplot as plt
import networkx as nx

# Membuat Graph ERD
G = nx.DiGraph()

# Menambahkan entitas
entities = ["User", "Journal", "Meditation", "Forum", "Mental Test", "Notification", "Psychologist"]
for entity in entities:
    G.add_node(entity, shape="rectangle")

# Menambahkan hubungan antar entitas
relations = [
    ("User", "Journal"), ("User", "Meditation"), ("User", "Forum"), ("User", "Mental Test"),
    ("User", "Notification"), ("User", "Psychologist"), ("Psychologist", "Forum"), ("Psychologist", "Mental Test")
]

for relation in relations:
    G.add_edge(*relation)

# Plot Diagram ERD
plt.figure(figsize=(10, 6))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color="lightblue", node_size=3000, edge_color="gray", font_size=10, font_weight="bold")
plt.title("Entity Relationship Diagram (ERD) - Love Yourself", fontsize=12)
plt.savefig("/mnt/data/ERD_Love_Yourself.png")  # Simpan sebagai file
plt.show()
```

```python
from graphviz import Digraph

# Membuat diagram ERD
erd = Digraph('ERD_Love_Yourself', filename='ERD_Love_Yourself', format='png')

# Menambahkan entitas
erd.node('User', 'User\n(UserID, Name, Email, Password)')
erd.node('Journal', 'Journal\n(JournalID, UserID, Entry, Date)')
erd.node('Meditation', 'Meditation\n(MeditationID, UserID, Duration, Type)')
erd.node('Forum', 'Forum\n(PostID, UserID, Content, Date)')
erd.node('MentalTest', 'Mental Test\n(TestID, UserID, Result, Date)')
erd.node('Notification', 'Notification\n(NotificationID, UserID, Message, Date)')
erd.node('Psychologist', 'Psychologist\n(PsychologistID, Name, Specialty)')

# Menambahkan relasi
erd.edge('User', 'Journal', label='writes')
erd.edge('User', 'Meditation', label='participates in')
erd.edge('User', 'Forum', label='posts in')
erd.edge('User', 'MentalTest', label='takes')
erd.edge('User', 'Notification', label='receives')
erd.edge('User', 'Psychologist', label='consults')
erd.edge('Psychologist', 'Forum', label='participates in')
erd.edge('Psychologist', 'MentalTest', label='creates')

# Menyimpan file
erd.render('/mnt/data/ERD_Love_Yourself')

# Menampilkan ERD
erd.view()
```

```python
import matplotlib.pyplot as plt
import networkx as nx

# Membuat diagram konteks sederhana
G = nx.DiGraph()

# Menambahkan node (Entitas dan Sistem)
G.add_node("Pengguna", shape="rectangle", style="filled", fillcolor="lightblue")
G.add_node("Aplikasi Love Yourself", shape="rectangle", style="filled", fillcolor="lightgreen")
G.add_node("Server Database", shape="rectangle", style="filled", fillcolor="lightgrey")

# Menambahkan edges (Aliran data)
G.add_edges_from([
    ("Pengguna", "Aplikasi Love Yourself", {"label": "Input Data, Permintaan Layanan"}),
    ("Aplikasi Love Yourself", "Server Database", {"label": "Simpan & Ambil Data"}),
    ("Server Database", "Aplikasi Love Yourself", {"label": "Balikan Data"}),
    ("Aplikasi Love Yourself", "Pengguna", {"label": "Tampilkan Informasi"}),
])

# Menggambar diagram
plt.figure(figsize=(8, 5))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color=['lightblue', 'lightgreen', 'lightgrey'], node_size=3000, font_size=10, edge_color="gray")
edge_labels = {(u, v): d["label"] for u, v, d in G.edges(data=True)}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=8)
plt.title("Diagram Konteks - Aplikasi Love Yourself")
plt.show()
```
