```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('aliandoo.jpg')
```

```python
watermark = np.zeros_like(image)
cv2.putText(watermark, 'KAU TERINDAH BY ILLY', (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)
```

```python
alpha = 0.5  # Transparansi watermark
watermarked_image = cv2.addWeighted(image, 1, watermark, alpha, 0)
```

```python
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Watermarked Image')
plt.imshow(cv2.cvtColor(watermarked_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.show()
```
