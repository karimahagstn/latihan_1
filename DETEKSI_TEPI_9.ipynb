```python
from google.colab import files
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.filters import roberts, prewitt

# Upload gambar
uploaded = files.upload()

# Baca gambar dan konversi ke grayscale
for file_name in uploaded.keys():
    image = cv2.imread(file_name, cv2.IMREAD_GRAYSCALE)

# Tampilkan gambar asli
plt.imshow(image, cmap="gray")
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
# Roberts Edge Detection
roberts_edges = roberts(image)

# Tampilkan hasil
plt.imshow(roberts_edges, cmap="gray")
plt.title("Roberts Edge Detection")
plt.axis("off")
plt.show()
```

```python
# Prewitt Edge Detection
prewitt_edges = prewitt(image)

# Tampilkan hasil
plt.imshow(prewitt_edges, cmap="gray")
plt.title("Prewitt Edge Detection")
plt.axis("off")
plt.show()
```

```python
# Sobel Edge Detection
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)  # Gradien x
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)  # Gradien y
sobel_edges = cv2.magnitude(sobel_x, sobel_y)

# Tampilkan hasil
plt.imshow(sobel_edges, cmap="gray")
plt.title("Sobel Edge Detection")
plt.axis("off")
plt.show()
```

```python
# Canny Edge Detection
canny_edges = cv2.Canny(image, 100, 200)

# Tampilkan hasil
plt.imshow(canny_edges, cmap="gray")
plt.title("Canny Edge Detection")
plt.axis("off")
plt.show()
```

```python
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.ravel()

axes[0].imshow(image, cmap="gray")
axes[0].set_title("Original Image")
axes[0].axis("off")

axes[1].imshow(roberts_edges, cmap="gray")
axes[1].set_title("Roberts")
axes[1].axis("off")

axes[2].imshow(prewitt_edges, cmap="gray")
axes[2].set_title("Prewitt")
axes[2].axis("off")

axes[3].imshow(sobel_edges, cmap="gray")
axes[3].set_title("Sobel")
axes[3].axis("off")

axes[4].imshow(canny_edges, cmap="gray")
axes[4].set_title("Canny")
axes[4].axis("off")

plt.tight_layout()
plt.show()
```
