```python
# Import library yang diperlukan
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris  # Dataset Iris
from sklearn.cluster import KMeans

# Memuat dataset Iris
iris = load_iris()
X = iris.data

# Menentukan jumlah cluster yang akan dicoba
K_values = [5, 6, 7]

# Membuat subplots untuk membandingkan hasil clustering
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for i, K in enumerate(K_values):
    # Menjalankan K-Means
    kmeans = KMeans(n_clusters=K, random_state=42, n_init=10)
    kmeans.fit(X)
    y_kmeans = kmeans.predict(X)

    # Scatter plot untuk hasil clustering
    axes[i].scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

    # Menampilkan pusat cluster (centroid)
    centers = kmeans.cluster_centers_
    axes[i].scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker="X", label="Centroids")

    # Menyesuaikan tampilan
    axes[i].set_xlabel(iris.feature_names[0])
    axes[i].set_ylabel(iris.feature_names[1])
    axes[i].set_title(f'K-Means Clustering (K={K})')
    axes[i].legend()

plt.tight_layout()
plt.show()
```
