```python
df['Start'] = pd.to_datetime(df['Time Start'].str.strip())
df['Finish'] = pd.to_datetime(df['Time Finish'].str.strip())
```

```python
import plotly.express as px

fig = px.timeline(df, x_start="Start", x_end="Finish", y="User", color="Laboratory",
                  hover_data=["Course", "Major", "Group"])
fig.update_yaxes(categoryorder="total ascending")
fig.update_layout(title='Jadwal Penggunaan Laboratorium (Gantt Chart)', height=600)
fig.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Ekstrak hari dan jam dari waktu mulai
df['Hari'] = df['Start'].dt.day_name()
df['Jam'] = df['Start'].dt.hour

# Buat pivot table: jumlah praktikum per hari dan jam
pivot_table = df.pivot_table(index='Hari', columns='Jam', aggfunc='size', fill_value=0)

# Urutkan hari sesuai urutan kalender
hari_urut = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
pivot_table = pivot_table.reindex(hari_urut)

# Plot heatmap
plt.figure(figsize=(15, 6))
sns.heatmap(pivot_table, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Heatmap Jumlah Praktikum per Hari dan Jam')
plt.xlabel('Jam (24-jam)')
plt.ylabel('Hari')
plt.tight_layout()
plt.show()
```
