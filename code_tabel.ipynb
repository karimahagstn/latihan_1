```python
import matplotlib.pyplot as plt
from matplotlib.table import Table

def create_ethereum_comparison_table_png(filename='ethereum_comparison.png'):
    data = [
        ['TPS', '1530', 'Hingga ribuan', 'Peningkatan throughput'],
        ['Latensi', 'Tinggi', 'Rendah', 'Respons lebih cepat'],
        ['Beban Node', 'Tinggi', 'Terdistribusi', 'Operasional lebih efisien'],
        ['Keamanan', 'Terpusat', 'Terdesentralisasi', 'Beacon chain & validator'],
        ['Interoperabilitas', 'Tidak ada', 'Terbatas', 'Tantangan protokol'],
    ]
    columns = ['Aspek', 'Ethereum 1.0', 'Ethereum 2.0 + Sharding', 'Dampak']

    fig, ax = plt.subplots(figsize=(8, 3.5))
    ax.set_axis_off()

    # Create table
    table = Table(ax, bbox=[0, 0, 1, 1])

    nrows = len(data) + 1
    ncols = len(columns)

    # Cell colors
    header_color = '#825affad'  # semi-transparent purple
    aspect_col_color = '#e2c044'  # yellowish
    eth1_color = '#f0a500'  # orange
    eth2_color = '#4dd599'  # turquoise
    impact_color = '#00c8ff'  # bright blue
    cell_bg = '#1e1e1e'  # dark background
    cell_bg_alt = '#29293f'  # alternate row background
    text_color = 'white'

    # Add header row
    for col in range(ncols):
        table.add_cell(0, col, 1/ncols, 0.15, text=columns[col], loc='center',
                       facecolor=header_color, edgecolor='black',
                       text_props=dict(weight='bold', color=text_color, fontsize=12))

    # Add data rows
    for row in range(1, nrows):
        for col in range(ncols):
            cell_text = data[row-1][col-1] if col > 0 else data[row-1][0]
            # Calculate which text to use
            if col == 0:
                cell_text = data[row-1][0]
                face = cell_bg
                color = aspect_col_color
                weight = 'bold'
            elif col == 1:
                cell_text = data[row-1][1]
                face = cell_bg
                color = eth1_color
                weight = 'bold'
            elif col == 2:
                cell_text = data[row-1][2]
                face = cell_bg
                color = eth2_color
                weight = 'bold'
            else:
                cell_text = data[row-1][3]
                face = cell_bg
                color = impact_color
                weight = 'bold'

            # Alternate row bg for readability
            if row % 2 == 0:
                face = cell_bg_alt

            table.add_cell(row, col, 1/ncols, 0.15, text=cell_text, loc='center',
                           facecolor=face, edgecolor='black',
                           text_props=dict(color=color, weight=weight, fontsize=11))

    # Row and col labels off
    ax.add_table(table)

    # Set tight layout and save
    plt.tight_layout()
    plt.savefig(filename, dpi=300, bbox_inches='tight', facecolor='#121212')
    plt.close()
    print(f"Generated table PNG saved as: {filename}")

if __name__ == '__main__':
    create_ethereum_comparison_table_png()

```
